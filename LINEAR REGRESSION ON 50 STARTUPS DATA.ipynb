### IMPORTING NECESSARY LIBRARIES


```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

#### READING THE FILE 

```python
df=pd.read_csv('50_Startups.csv')
```

```python
df.head(1)
```

SPLITING THE DATA IN X AND Y

```python
x=df.iloc[:,:-1]
```

```python
y=df.iloc[:,-1]
```

##### CHECKING THE SHAPE

```python
print(x.shape)
print(y.shape)
```

##### CREATING THE DUMMY DATAFRAME

```python
s=pd.get_dummies(x['State'])
```

```python
s.head(1)
```

##### DROPPING THE STATE COLUMNS

```python
x.drop('State', axis=1, inplace=True)
```

#### CONCATINATING BOTH THE DATAFRAMES

```python
x=pd.concat([x,s],axis=1)
```

#### VIEWING THE HEAD

```python
x.head(1)
```

#### PLOTTING THE PAIRPLOT FROM SEABORN LIBRARY

```python
sns.pairplot(x)
```

##### IMPORTING THE TRAIN_TEST_SPLIT 

```python
from sklearn.model_selection import train_test_split
```

##### SPLITTING THE WHOLE DATASET INTO TRAIN AND TEST DATASETS

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

##### LINEAR REGRESSION MODELLING

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

##### PREDICTING 

```python
y_pred=lr.predict(x_test)
```

##### IMPORTING THE R2 SCORE

```python
from sklearn.metrics import r2_score
```

```python
r_square=r2_score(y_test,y_pred)
```

```python
print(r_square)
```

###### THE VALUE OF THE R2 SCORE IS 0.978
